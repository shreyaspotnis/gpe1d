# Oscillations
In this example, we will setup the GPE equation to find the ground state of a BEC in a harmonic trap and then look at center of mass and breathing mode oscillations.

First, import the modules we need and set matplotlib font size to something bigger than the default

```python
%matplotlib inline
import numpy as np
import numpy.fft as fft
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams.update({'font.size': 14, 'font.family': 'serif'})

from gpe1d.gpe1d import gpe1d_python
```

Now let's define some constants that we will be using

```python
# Fundamental constants
pi = 3.1415926
hbar = 1.054571628e-34  # (SI) Reduced Planck's constant
h = hbar*2*pi           # (SI) Planck's constant
kb = 1.3806e-23         # (SI) Boltzman constant

# Rubidium constants
m_rb = 1.443e-25         # (kg) Mass of Rubiduim87
a_bg = 5.1e-09           # (m) Rubidium87 background scattering length
```

Defining some parameters of the system we are simulating

```python
fx = 20          # (Hz) Initial trapping frequency along the direction in which we are studying
fy = fz = 200    # (Hz) Transverse trapping frequencies
n_atoms = 100e3  # number of atoms in the trap
```

Next, we will define some parameters which are relevant for the simulation

```python
size = 12
n_x = 2 ** size         # since we are using fft for the simulation, we keep our grid size to be a power of 2
length_scale = 5e-6    # typical length scale of the problem, used to non-dimensionalize the GPE equation
d_x = 1.0/32.0          # grid spacing in non-dimensionalized units
time_scale = 1.0/fx    # we use the longitudinal trapping period as the time scale for non-dimensionalization
d_t = 1.0/5000.0        # time spacing in non-dim units
n_t = int(2.0/d_t)       # number of time steps, we choose 2 periods
n_t_store = 40           # number of snapshots in time of the wavefunction to save
n_t_imag = int(5.0/d_t)  # number of imaginary time steps to take while finding the ground state wavefunction
```

This seems like it's never going to end, but we need to calculate some additional quantities based on the ones defined previously

```python
u_0 = 4.0*pi*hbar**2*a_bg/m_rb  # two particle interaction energy

# Calculate non-interacting gaussian widths
wx = (hbar/(m_rb*2.0*pi*fx))**0.5
wy = (hbar/(m_rb*2.0*pi*fy))**0.5
wz = (hbar/(m_rb*2.0*pi*fz))**0.5

energy_scale = m_rb*length_scale**2/time_scale**2 # (kg m^2/s^2)
delta = 4.0*pi*a_bg*n_atoms/wx
epsilon = (wx/length_scale)**2
gamma_y = fy/fx
gamma_z = fz/fx

# Effective interaction term when reducing the 3d GPE to 1d GPE for strong interactions
kappa_eff_1d =((delta*gamma_y*gamma_z)**(3.0/5.0)*epsilon**1.5*pi/9.0*(15.0/4.0/pi)**(8.0/5.0))
```

X grid, add info here

```python
X = (np.arange(0,n_x)-(n_x-1)/2.0)*d_x
X_real = X * length_scale
```

Initialize the trapping potential

```python
U_0 = 0.5*X**2  # we've chosen our time and energy scales wisely
# U_0 = 0.5 * m_rb * (2.0 * pi * fx) ** 2 * X_real ** 2 / energy_scale
```

Calculate the approximate ground state wavefunction based on the Thomas-Fermi approximation

```python
# The approximate ground state wavefuction for the 1d gpe
mu_1d = (3.0 / 2.0 * kappa_eff_1d) ** (2.0 / 3.0) /2.0  # chemical potential
r_tf_x = (2.0 * mu_1d) ** 0.5  # Thomas fermi radius in the x-direction

psi_1d_tf = (mu_1d - U_0) / kappa_eff_1d
psi_1d_tf = ((psi_1d_tf > 0) * psi_1d_tf) ** 0.5

def Prob(psi):
    return np.real(psi*np.conj(psi))

plt.plot(X_real * 1e6, Prob(psi_1d_tf), 'black')
plt.xlabel('$X(\mu m)$')
plt.ylabel('$|\psi|^2$')
plt.show()
```

Propagate through imaginary time to find the ground state

```python
(K, T_store, psi_ground, _) = gpe1d_python(epsilon, kappa_eff_1d, n_t_imag, d_t, X, U_0, psi_1d_tf, n_t_store, imag_time=1)
```

The ground state found numerically is very close to our Thomas Fermi approximation. The difference is apparent only at the edges, where the actual ground state does not have the sharp kink that the Thomas Fermi approximation has.

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
fig.set_size_inches((12, 4))

ax[0].plot(X_real * 1e6, Prob(psi_1d_tf), 'r')
ax[0].plot(X_real * 1e6, Prob(psi_ground), 'black')
ax[0].set_xlabel('$X(\mu m)$')
ax[0].set_ylabel('$|\psi|^2$')

ax[1].plot(X_real * 1e6, Prob(psi_ground) - Prob(psi_1d_tf), 'black')
ax[1].set_xlabel('$X(\mu m)$')
ax[1].set_ylabel('$|\psi|^2 - |\psi_{TF}|^2$')

fig.tight_layout()
```

Next, we will give the ground state a velocity kick 

```python
v_kick = 1e-3  # 4mm/s
k_kick = m_rb * v_kick / hbar * length_scale  # non-dimensionalized k-vector
psi_initial = psi_ground * np.exp(1j * k_kick * X)
```

```python
(K, T_store, psi_osc, _) = gpe1d_python(epsilon, kappa_eff_1d, n_t_imag, d_t, X, U_0, psi_initial, n_t_store, imag_time=0)
```

Let's plot the expectation value of position and velocity of our wavepacket

```python
d_k = K[1]-K[0]
K_real = K / length_scale
T_real = T_store * time_scale

psi2_osc = Prob(psi_osc)

psi_osc_k = fft.fft(psi_osc)  # psi in k space
psi_osc_k *= (d_x / d_k / n_x) ** 0.5  # normalization

# find expectation value of x, v
x_ex = np.sum(psi2_osc * X, axis=1) * length_scale * d_x
```

```python
plt.plot(T_real * 1e3, x_ex * 1e6)
plt.show()
```
